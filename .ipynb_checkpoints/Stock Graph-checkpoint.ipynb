```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
%matplotlib inline
```

```python
APPL_STOCK = pd.read_csv('AAP.csv')
APPL_STOCK.head()
```

```python
plt.tight_layout()
APPL_STOCK.plot(figsize=(15,5),x='Date', y='Close',)
```

```python
fig = plt.figure(figsize=(10, 6))
plt.xlabel('Date',fontsize=20)
plt.ylabel('Value',fontsize=20)
plt.title('Apple Stock',fontsize=20)
```

```python
def animate(i):
    data = APPL_STOCK['Close'].iloc[:i]
    return data
```
